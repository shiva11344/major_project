#### Fraud dataset which has 400 entries from different users  

```python
df0 = pd.read_csv('C:\\Users\\91998\\Desktop\\MI project\\sf.csv')
df0.head()
```

#### Labelling 0 to frauddata

```python
label =[]
label=[0]*400
df0['result']=label
df0
```

#### Dataset which has 1000  entries took at different sessions from genuine user

```python
df1 = pd.read_csv('C:\\Users\\91998\\Desktop\\MI project\\s.csv')
df1.head()
```

#### Labelling 1 to genuine data

```python
label =[]
label=[1]*1000
df1['result']=label
df1
```

#### Combining two data sets


```python
df=df1.append(df0, ignore_index = True) 
df
```

```python
#shuffling all the rows of final dataset
df= df.sample(frac=1)
```

```python
df
```

```python
df.to_csv('finaldataset.csv')
```

```python
df=df.set_index('Name')
df
```

#### Splitting data

```python
y= df['result']
x= df.loc[:, df.columns != 'result']
```

```python
print(len(x))
print(len(y))
```

```python
import keras
import tensorflow as tf
from keras.preprocessing.sequence import TimeseriesGenerator
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.utils import to_categorical
from keras.optimizers import Adam
```

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
x_train
```

#### Reshaping into 3D to input into LSTM layer


```python
x_train = x_train.values.reshape(1260, 1, 21)
```

```python
x_test  = x_test.values.reshape(140,1,21)
```

```python
x_train.shape
```

#### Training the data

```python
model = Sequential()
model.add(LSTM(80, activation='softmax', input_shape=(1, 21)))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse',metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(x_train, y_train, epochs=20, validation_data=(x_test,y_test), batch_size=3)
```

```python
model.summary()
```

#### Validation 

```python
predictions = model.predict(x_test)
predictions
```

```python
import numpy as ro
pred=ro.round(predictions,0)
```
