```python
df2 = pd.read_csv('shiva_timing_2.csv')
df2
```

```python
df3 = pd.read_csv('shiva_fraud.csv')
df3
```

```python
dataframe=pd.concat([df1,df2,df3], axis=0)
```

```python
dataframe = dataframe.sample(frac=1).reset_index(drop=True)
```

```python
dataframe=dataframe.loc[:,'a':'Result']
```

```python
# Load libraries
import pandas
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
array = dataframe.values
X = array[:,0:-1]
Y = array[:,-1]
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

```python
Y_train.shape
```

```python
X_train.shape
```

```python
# Test options and evaluation metric
seed = 7
scoring = 'accuracy'
print(X_train.shape,Y_train.shape)
```

```python
from sklearn import svm
svc = svm.SVC(kernel='linear')
svc.fit(X_train,Y_train)
```

```python
y_pred=svc.predict(X_validation)
```

```python
y_pred
```

```python
from sklearn import metrics 
print("svm model accuracy:", metrics.accuracy_score(y_pred, Y_validation))
```

```python
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors=3) 
knn.fit(X_train, Y_train) 
```

```python
y_pred = knn.predict(X_validation) 
  
# comparing actual response values (y_test) with predicted response values (y_pred) 
from sklearn import metrics 
print("kNN model accuracy:", metrics.accuracy_score(Y_validation, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```
