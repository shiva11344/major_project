#### Splitting data

```python
result_data = df['result'].values
result_data = result_data.reshape(-1,1)
split_percentage = 0.80
split = int(split_percentage*len(result_data))
result_data[0:1000]
```

```python
result_train = result_data[:split]
result_test = result_data[split:]
data_train = df[:split]
data_test = df[split:]
```

```python
print(len(result_train))
print(len(result_test))
print(len(data_train))
print(len(data_test))
result_train.shape
data_train.shape
```

```python
import keras
import tensorflow as tf
from keras.preprocessing.sequence import TimeseriesGenerator
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
lb = 1
train_gen = TimeseriesGenerator(data_train, result_train, length=lb, batch_size=1)     
test_gen = TimeseriesGenerator(result_test, result_test, length=lb, batch_size=1)
```

```python
mod = Sequential()
mod.add(
    LSTM(10,
    activation='softmax',
    input_shape=(1120,23))
)

mod.add(Dense(1))
mod.add(Dense(1))
mod.compile(optimizer='adam', loss='mse')
```

```python
mod.summary()
```

```python
epoch = 20
```

```python
mod.fit_generator(train_gen, epochs=epoch, verbose=1)
```

```python
prediction=mod.predict_generator(test_gen)
print(prediction)
```

```python
import numpy as np
pred=np.round(prediction,0)
print(pred)
```
