```python
def legend_without_duplicate_labels(ax):
    handles, labels = ax.get_legend_handles_labels()
    unique = [(h, l) for i, (h, l) in enumerate(zip(handles, labels)) if l not in labels[:i]]
    ax.legend(*zip(*unique))
for i in range(0,4):
    df_y = df.iloc[i,3:].values
    df_y = df_y.tolist()
    plt.plot(headers, df_y,label="Genuine",color='green')
for i in range(0,4):
    df2_y = df2.iloc[i,3:].values
    df2_y = df2_y.tolist()
    plt.plot(headers, df2_y,label="Fraud",color='Red')

plt.xlabel('H,DD,UD,UU Keys Notation')
plt.ylabel('Keystroke Timings')
plt.legend(loc="upper left")

def update_prop(handle, orig):
    handle.update_from(orig)
    handle.set_marker("")
    
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

fig = plt.gcf()
fig.set_size_inches(18.0, 7, forward=True)

plt.show()
```
